```python
# Importing libraries.

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Opening the image and showing it.

src = cv.imread('./images/images.png', cv.IMREAD_COLOR)
plt.imshow(cv.cvtColor(src, cv.COLOR_BGR2RGB))
```

```python
# Grayscaling the image and showing it.

gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
# gray = cv.medianBlur(gray, 5)

plt.imshow(gray, cmap='gray')
```

```python
# Getting the width and height of image.

height = gray.shape[0]
width = gray.shape[1]
print("{} X {}".format(height, width))
```

```python
circles = cv.HoughCircles(gray, cv.HOUGH_GRADIENT, 1, 10,
                  param1=30, param2=60,
                  minRadius=40, maxRadius=400)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        center = (i[0], i[1])
        # circle center
        cv.circle(src, center, 1, (0, 255, 0), 1)
        # circle outline
        radius = i[2]
        print(center, radius)
        cv.circle(src, center, radius, (0, 255, 0), 2)
        
plt.imshow(src)
```

```python
# Applying canny edge detection.

edges = cv.Canny(gray, 80, 100)
plt.imshow(edges)
```

```python
# Applying circle detection on edged image

circles = cv.HoughCircles(edges, cv.HOUGH_GRADIENT, 1, 10,
                  param1=30, param2=60,
                  minRadius=40, maxRadius=400)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        center = (i[0], i[1])
        # circle center
        cv.circle(src, center, 1, (0, 255, 0), 1)
        # circle outline
        radius = i[2]
        print(center, radius)
        cv.circle(src, center, radius, (0, 255, 0), 2)
        
plt.imshow(src)
```
